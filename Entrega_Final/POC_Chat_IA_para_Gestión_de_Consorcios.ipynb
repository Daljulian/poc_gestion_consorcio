```python
BASE_DE_CONOCIMIENTO = {
    "expensas": "Expensas de Septiembre: 15.000 ARS. Vencimiento: 25/09/2025.",
    "reglamento": "Se prohbe el uso de la pileta despus de las 22:00 hs. Multas por ruidos segn art. 15.",
    "plomero": "Plomero autorizado: Juan Prez, tel: 11-1234-5678.",
    "electricista": "Electricista de confianza: Pedro Gmez, tel: 11-9876-5432."
}
```

```python
#Dataset sinttico para entrenar un clasificador rpido (ejemplo)
import pandas as pd

# Ejemplos sintticos (ampliables)
data = [
    ("Quiero pagar las expensas", "pagos"),
    ("Cundo vencen las expensas?", "pagos"),
    ("Tengo retraso en el pago", "pagos"),
    ("Se rompi la canilla en mi cocina", "reparaciones"),
    ("Hay una prdida de agua en 3B", "reparaciones"),
    ("El ascensor no funciona", "reparaciones"),
    ("Necesito el contrato de alquiler", "documentacion"),
    ("Dnde est el reglamento?", "documentacion"),
    ("Quiero alquilar la unidad 2A", "ofertas"),
    ("Hay unidades disponibles?", "ofertas"),
    ("Hola", "general"),
    ("Gracias", "general"),
    ("Quin gan el partido?", "general")
]
# Este df terminaria siendo la base de conocimiento.
df = pd.DataFrame(data, columns=["texto", "categoria"])
df
```

```python
#Entrenar un clasificador ML simple (TF-IDF + LogisticRegression)

from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LogisticRegression

# Pipeline sencillo
pipeline = Pipeline([
    ("tfidf", TfidfVectorizer()),
    ("clf", LogisticRegression(max_iter=1000))
])

X = df["texto"]
y = df["categoria"]

pipeline.fit(X, y)

# Prueba rpida
test_msgs = [
    "La luz del pasillo no funciona",
    "Cmo pago las expensas?",
    "Necesito ver mi contrato",
    "Busco un plomero"
]

for m in test_msgs:
    print(m, "->", pipeline.predict([m])[0])
```

```python
#Motor de clasificacin combinado: reglas + ML
#La idea: primer filtro rule-based (palabras clave) y si no hay match usar el clasificador ML.

import re

# Diccionario de palabras clave
keywords = {
    "pagos": ["expensa", "pagar", "venc", "saldo", "cuota"],
    "reparaciones": ["rotur", "perdida", "fuga", "ascensor", "desperfect", "plom"],
    "documentacion": ["contrato", "document", "reglamento", "escritura"],
    "ofertas": ["alquil", "venta", "unidad", "oferta"]
}

def clasificador_combinado(texto: str) -> str:
    txt = texto.lower()
    # regla simple
    for cat, keys in keywords.items():
        if any(re.search(r"\b" + k, txt) for k in keys):
            return cat
    # fallback al ML
    return pipeline.predict([texto])[0]

# test
for m in test_msgs:
    print(m, "->", clasificador_combinado(m))
```

```python
#Generador de respuestas (simulacin) y hook para IA real
#generar_respuesta_simulada usa la base de conocimiento y plantillas.
#generar_respuesta_api es un placeholder que, si defins las claves, puede usarse con Google Generative AI o OpenAI

import random

RESPUESTAS_TEMPL = {
    "pagos": [
        lambda _: BASE_DE_CONOCIMIENTO["expensas"],
        lambda _: "Puede abonar las expensas a travs del link de pago o en la administracin."
    ],
    "reparaciones": [
        lambda _: "Se registr su reclamo. Indique nombre, nmero de unidad y descripcin detallada.",
        lambda _: "Un tcnico ser asignado en breve; por favor provea una foto si es posible."
    ],
    "documentacion": [
        lambda _: "La documentacin solicitada puede retirarse en la oficina o solicitarse por correo.",
        lambda _: "El reglamento interno est disponible: " + BASE_DE_CONOCIMIENTO["reglamento"]
    ],
    "ofertas": [
        lambda _: "Hay unidades disponibles. Desea que le envie los detalles por email?",
        lambda _: "Se registr su inters. El rea comercial lo contactar."
    ],
    "general": [
        lambda _: "Su consulta fue registrada. Un administrador la revisar.",
        lambda _: "Por favor, ample su consulta para que pueda asistirlo mejor."
    ]
}

def generar_respuesta_simulada(categoria, texto_usuario):
    plantilla = random.choice(RESPUESTAS_TEMPL.get(categoria, RESPUESTAS_TEMPL["general"]))
    return plantilla(texto_usuario)

# Placeholder para integrar API real (ejemplo con Google Generative AI)
def generar_respuesta_api(mensaje_usuario):
    # Ejemplo (no ejecuta): si tens GOOGLE_API_KEY configurada, pods descomentar
    # y adaptar al SDK que uses (google.generativeai o openai).
    #
    # import google.generativeai as genai
    # genai.configure(api_key=GOOGLE_API_KEY)
    # model = genai.GenerativeModel('gemini-pro')
    # prompt = f"Contexto: {BASE_DE_CONOCIMIENTO}\nUsuario: {mensaje_usuario}\nBot:"
    # response = model.generate_content(prompt)
    # return response.text
    #
    # Por ahora devolvemos None para indicar que no hay API configurada.
    return None
```

```python
#Interfaz interactiva en consola (funciona en Colab con input())

import csv
from datetime import datetime

LOG_CSV = "interactions_log.csv"

# guardar cabecera
with open(LOG_CSV, "w", newline="", encoding="utf-8") as f:
    writer = csv.DictWriter(f, fieldnames=["ts", "usuario", "mensaje", "categoria", "respuesta", "motor"])
    writer.writeheader()

def registrar_log(usuario, mensaje, categoria, respuesta, motor="simulado"):
    with open(LOG_CSV, "a", newline="", encoding="utf-8") as f:
        writer = csv.DictWriter(f, fieldnames=["ts", "usuario", "mensaje", "categoria", "respuesta", "motor"])
        writer.writerow({
            "ts": datetime.utcnow().isoformat(),
            "usuario": usuario,
            "mensaje": mensaje,
            "categoria": categoria,
            "respuesta": respuesta,
            "motor": motor
        })

print("=== POC Chat IA - Simulacin interactiva ===")
usuario = input("Ingrese nombre o rol (ej: Propietario, Inquilino): ").strip()

print("Escriba 'salir' para finalizar. Escriba 'modo api' para probar hook de API (si est configurada).")

modo_api = False
while True:
    texto = input("> ").strip()
    if texto.lower() == "salir":
        break
    if texto.lower() == "modo api":
        modo_api = not modo_api
        print("Modo API activado." if modo_api else "Modo API desactivado.")
        continue

    categoria = clasificador_combinado(texto)
    respuesta_api = None
    if modo_api and OPENAI_API_KEY:
        respuesta_api = generar_respuesta_api(texto)

    if respuesta_api:
        respuesta = respuesta_api
        motor = "api"
    else:
        respuesta = generar_respuesta_simulada(categoria, texto)
        motor = "simulado"

    print(f"[ConsorcioBot - {categoria}] {respuesta}\n")
    registrar_log(usuario, texto, categoria, respuesta, motor)

print("Sesin finalizada. Logs guardados en", LOG_CSV)
```

# Sobre el Prototipo Avanzado


 Metodologa

El proyecto se llevar a cabo bajo un enfoque iterativo e incremental, comenzando con un prototipo bsico (Gemini API) para validar la capacidad de la IA de generar respuestas tiles, y evolucionando hacia un prototipo avanzado (Twilio + Supabase) que integra comunicacin real y almacenamiento persistente.

Los procedimientos implementados incluyen:

Diseo modular: cada componente (clasificador, IA, comunicacin, base de datos) se desarrolla de forma independiente para facilitar pruebas y mejoras.

Validacin progresiva: primero se prueban respuestas en entorno controlado (Google Colab) y luego se validan en un flujo real con Twilio.

Retrieval-Augmented Generation (RAG): se gua a la IA con informacin de la base de conocimiento para evitar respuestas incorrectas o inventadas.

 Justificacin: este enfoque permite reducir riesgos y garantizar resultados tangibles en cada fase, asegurando que los objetivos se cumplen antes de pasar al siguiente nivel.

 Herramientas y Tecnologas

Google Gemini API: motor de IA para generacin de respuestas.

Twilio: comunicacin por WhatsApp/SMS con usuarios reales.

Supabase: base de datos relacional en la nube, con autenticacin y almacenamiento dinmico.

Flask / FastAPI: servidor web para manejar la lgica y conectar servicios.

Python + Scikit-learn: clasificacin de mensajes con TF-IDF + Logistic Regression.

dotenv: manejo seguro de claves de API.

Tcnicas de Fast Prompting utilizadas

Role prompting: se define claramente el rol de la IA (asistente virtual de consorcio).

Instruction prompting: se establecen reglas explcitas (no inventes soluciones, responde solo sobre el consorcio).

Context prompting: se aporta informacin del clasificador y la base de conocimiento como contexto.

Guardrails prompting: se agregan instrucciones de seguridad para evitar respuestas fuera de dominio.

Prompt colaborativo para el desarrollo:

Constraint-Based Prompting: Este mtodo implica establecer reglas y limitaciones claras para la respuesta para el desarrollo de los prototipos.

AI-in-the-loop: (IA en el bucle) o, de forma ms especfica, "iterative prompting" (prompting iterativo)

Role Prompting: Asignar un rol o persona especfica al modelo puede cambiar drsticamente el tono y el estilo de su respuesta. Al indicarle al modelo que acte como un experto

Few-Shot Prompting: tcnica consiste en incluir uno o ms ejemplos del formato de entrada y salida deseado dentro del prompt.


**Justificacin**: estas tcnicas aseguran respuestas relevantes, seguras y consistentes, alineadas con los objetivos del proyecto.

```python
#Implementacin

#Ejemplo de cdigo (simplificado, Gemini API)

import os
from dotenv import load_dotenv
import google.generativeai as genai

# Cargar variables de entorno
load_dotenv()
GOOGLE_API_KEY = os.getenv("GOOGLE_API_KEY")

# Configuracin de la API
genai.configure(api_key=GOOGLE_API_KEY)
model = genai.GenerativeModel('gemini-pro')
chat = model.start_chat(history=[])

BASE_DE_CONOCIMIENTO = {
    "pagos": "Las expensas vencen el da 10 de cada mes.",
    "documentacin": "Los reglamentos pueden solicitarse en administracin.",
    "reparaciones": "Debe detallarse el problema para derivar al proveedor correspondiente."
}

def generar_respuesta_ia(mensaje_usuario):
    prompt = f"""
    Eres un asistente virtual de consorcio.
    Solo respondes sobre expensas, documentacin o reclamos de reparaciones.
    Pregunta del usuario: "{mensaje_usuario}"
    """
    response = chat.send_message(prompt)
    return response.text

# Prueba
print(generar_respuesta_ia("Cundo vencen las expensas?"))
```

```python
#Ejemplo de cdigo (simplificado, Twilio + Supabase)

from flask import Flask, request
from twilio.twiml.messaging_response import MessagingResponse
from supabase import create_client

app = Flask(__name__)

SUPABASE_URL = os.getenv("SUPABASE_URL")
SUPABASE_KEY = os.getenv("SUPABASE_KEY")
supabase_client = create_client(SUPABASE_URL, SUPABASE_KEY)

@app.route("/sms", methods=['POST'])
def sms_reply():
    mensaje = request.form.get('Body')
    numero = request.form.get('From')

    # Generar respuesta con IA
    respuesta = generar_respuesta_ia(mensaje)

    # Registrar en Supabase
    supabase_client.table("reclamos").insert({
        "telefono": numero,
        "mensaje": mensaje,
        "respuesta": respuesta
    }).execute()

    # Responder al usuario
    resp = MessagingResponse()
    resp.message(respuesta)
    return str(resp)

if __name__ == "__main__":
    app.run(debug=True)
```
