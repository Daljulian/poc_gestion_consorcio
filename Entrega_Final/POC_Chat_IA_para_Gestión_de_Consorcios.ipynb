# POC: Chat IA para Gestión de Consorcios

**Resumen del Proyecto**

Este proyecto presenta una prueba de concepto (POC) de un chat de inteligencia artificial diseñado para asistir en la gestión diaria de consorcios e inmobiliarias. La herramienta centraliza consultas, reclamos y notificaciones, ofreciendo respuestas automáticas y clasificadas según la naturaleza de la solicitud.

El objetivo es validar la viabilidad técnica de un sistema conversacional capaz de reducir la carga administrativa, mejorar la comunicación entre propietarios, inquilinos, administradores y proveedores, y aportar transparencia en la gestión de los consorcios.

**Introducción**
**Presentación del problema**

La administración de consorcios e inmobiliarias enfrenta una gran saturación de notificaciones y consultas. Diariamente se reciben reclamos por desperfectos, solicitudes de reparación, dudas sobre pagos o documentación, y consultas sobre unidades disponibles. La gestión manual de estas interacciones provoca retrasos, errores y sobrecarga operativa para los administradores.

**Desarrollo de la solución**

La propuesta consiste en implementar un chat IA que automatice la atención de consultas y reclamos:

Clasifica automáticamente las consultas según su tipo (pagos, reclamos, documentación, ofertas).

Genera respuestas automáticas y contextuales basadas en un repositorio de información predefinido.

Canaliza solicitudes que requieren intervención humana (por ejemplo, asignación de reparaciones).

El sistema permite centralizar la comunicación y optimizar la atención de los usuarios, liberando recursos humanos para tareas de mayor valor agregado.

```python
BASE_DE_CONOCIMIENTO = {
    "expensas": "Expensas de Septiembre: 15.000 ARS. Vencimiento: 25/09/2025.",
    "reglamento": "Se prohíbe el uso de la pileta después de las 22:00 hs. Multas por ruidos según art. 15.",
    "plomero": "Plomero autorizado: Juan Pérez, tel: 11-1234-5678.",
    "electricista": "Electricista de confianza: Pedro Gómez, tel: 11-9876-5432."
}
```

```python
#Dataset sintético para entrenar un clasificador rápido (ejemplo)
import pandas as pd

# Ejemplos sintéticos (ampliables)
data = [
    ("Quiero pagar las expensas", "pagos"),
    ("¿Cuándo vencen las expensas?", "pagos"),
    ("Tengo retraso en el pago", "pagos"),
    ("Se rompió la canilla en mi cocina", "reparaciones"),
    ("Hay una pérdida de agua en 3B", "reparaciones"),
    ("El ascensor no funciona", "reparaciones"),
    ("Necesito el contrato de alquiler", "documentacion"),
    ("¿Dónde está el reglamento?", "documentacion"),
    ("Quiero alquilar la unidad 2A", "ofertas"),
    ("¿Hay unidades disponibles?", "ofertas"),
    ("Hola", "general"),
    ("Gracias", "general"),
    ("¿Quién ganó el partido?", "general")
]
# Este df terminaria siendo la base de conocimiento.
df = pd.DataFrame(data, columns=["texto", "categoria"])
df
```

```python
#Entrenar un clasificador ML simple (TF-IDF + LogisticRegression)

from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LogisticRegression

# Pipeline sencillo
pipeline = Pipeline([
    ("tfidf", TfidfVectorizer()),
    ("clf", LogisticRegression(max_iter=1000))
])

X = df["texto"]
y = df["categoria"]

pipeline.fit(X, y)

# Prueba rápida
test_msgs = [
    "La luz del pasillo no funciona",
    "¿Cómo pago las expensas?",
    "Necesito ver mi contrato",
    "Busco un plomero"
]

for m in test_msgs:
    print(m, "->", pipeline.predict([m])[0])
```

```python
#Motor de clasificación combinado: reglas + ML
#La idea: primer filtro rule-based (palabras clave) y si no hay match usar el clasificador ML.

import re

# Diccionario de palabras clave
keywords = {
    "pagos": ["expensa", "pagar", "venc", "saldo", "cuota"],
    "reparaciones": ["rotur", "perdida", "fuga", "ascensor", "desperfect", "plom"],
    "documentacion": ["contrato", "document", "reglamento", "escritura"],
    "ofertas": ["alquil", "venta", "unidad", "oferta"]
}

def clasificador_combinado(texto: str) -> str:
    txt = texto.lower()
    # regla simple
    for cat, keys in keywords.items():
        if any(re.search(r"\b" + k, txt) for k in keys):
            return cat
    # fallback al ML
    return pipeline.predict([texto])[0]

# test
for m in test_msgs:
    print(m, "->", clasificador_combinado(m))
```

```python
#Generador de respuestas (simulación) y hook para IA real
#generar_respuesta_simulada usa la base de conocimiento y plantillas.
#generar_respuesta_api es un placeholder que, si definís las claves, puede usarse con Google Generative AI o OpenAI

import random

RESPUESTAS_TEMPL = {
    "pagos": [
        lambda _: BASE_DE_CONOCIMIENTO["expensas"],
        lambda _: "Puede abonar las expensas a través del link de pago o en la administración."
    ],
    "reparaciones": [
        lambda _: "Se registró su reclamo. Indique nombre, número de unidad y descripción detallada.",
        lambda _: "Un técnico será asignado en breve; por favor provea una foto si es posible."
    ],
    "documentacion": [
        lambda _: "La documentación solicitada puede retirarse en la oficina o solicitarse por correo.",
        lambda _: "El reglamento interno está disponible: " + BASE_DE_CONOCIMIENTO["reglamento"]
    ],
    "ofertas": [
        lambda _: "Hay unidades disponibles. ¿Desea que le envie los detalles por email?",
        lambda _: "Se registró su interés. El área comercial lo contactará."
    ],
    "general": [
        lambda _: "Su consulta fue registrada. Un administrador la revisará.",
        lambda _: "Por favor, amplíe su consulta para que pueda asistirlo mejor."
    ]
}

def generar_respuesta_simulada(categoria, texto_usuario):
    plantilla = random.choice(RESPUESTAS_TEMPL.get(categoria, RESPUESTAS_TEMPL["general"]))
    return plantilla(texto_usuario)

# Placeholder para integrar API real (ejemplo con Google Generative AI)
def generar_respuesta_api(mensaje_usuario):
    # Ejemplo (no ejecuta): si tenés GOOGLE_API_KEY configurada, podés descomentar
    # y adaptar al SDK que uses (google.generativeai o openai).
    #
    # import google.generativeai as genai
    # genai.configure(api_key=GOOGLE_API_KEY)
    # model = genai.GenerativeModel('gemini-pro')
    # prompt = f"Contexto: {BASE_DE_CONOCIMIENTO}\nUsuario: {mensaje_usuario}\nBot:"
    # response = model.generate_content(prompt)
    # return response.text
    #
    # Por ahora devolvemos None para indicar que no hay API configurada.
    return None
```

```python
#Interfaz interactiva en consola (funciona en Colab con input())

import csv
from datetime import datetime

LOG_CSV = "interactions_log.csv"

# guardar cabecera
with open(LOG_CSV, "w", newline="", encoding="utf-8") as f:
    writer = csv.DictWriter(f, fieldnames=["ts", "usuario", "mensaje", "categoria", "respuesta", "motor"])
    writer.writeheader()

def registrar_log(usuario, mensaje, categoria, respuesta, motor="simulado"):
    with open(LOG_CSV, "a", newline="", encoding="utf-8") as f:
        writer = csv.DictWriter(f, fieldnames=["ts", "usuario", "mensaje", "categoria", "respuesta", "motor"])
        writer.writerow({
            "ts": datetime.utcnow().isoformat(),
            "usuario": usuario,
            "mensaje": mensaje,
            "categoria": categoria,
            "respuesta": respuesta,
            "motor": motor
        })

print("=== POC Chat IA - Simulación interactiva ===")
usuario = input("Ingrese nombre o rol (ej: Propietario, Inquilino): ").strip()

print("Escriba 'salir' para finalizar. Escriba 'modo api' para probar hook de API (si está configurada).")

modo_api = False
while True:
    texto = input("> ").strip()
    if texto.lower() == "salir":
        break
    if texto.lower() == "modo api":
        modo_api = not modo_api
        print("Modo API activado." if modo_api else "Modo API desactivado.")
        continue

    categoria = clasificador_combinado(texto)
    respuesta_api = None
    if modo_api and OPENAI_API_KEY:
        respuesta_api = generar_respuesta_api(texto)

    if respuesta_api:
        respuesta = respuesta_api
        motor = "api"
    else:
        respuesta = generar_respuesta_simulada(categoria, texto)
        motor = "simulado"

    print(f"[ConsorcioBot - {categoria}] {respuesta}\n")
    registrar_log(usuario, texto, categoria, respuesta, motor)

print("Sesión finalizada. Logs guardados en", LOG_CSV)
```

```python
#Base de conocimiento y clasificación

BASE_DE_CONOCIMIENTO = {
    "expensas": "Expensas de Septiembre: 15.000 ARS. Vencimiento: 25/09/2025.",
    "reglamento": "Se prohíbe el uso de la pileta después de las 22:00 hs. Multas por ruidos según art. 15.",
    "plomero": "Plomero autorizado: Juan Pérez, tel: 11-1234-5678.",
    "electricista": "Electricista de confianza: Pedro Gómez, tel: 11-9876-5432."
}

```

```python
#Dataset sintético para entrenar un clasificador rápido (ejemplo)
import pandas as pd

# Ejemplos sintéticos (ampliables)
data = [
    ("Quiero pagar las expensas", "pagos"),
    ("¿Cuándo vencen las expensas?", "pagos"),
    ("Tengo retraso en el pago", "pagos"),
    ("Se rompió la canilla en mi cocina", "reparaciones"),
    ("Hay una pérdida de agua en 3B", "reparaciones"),
    ("El ascensor no funciona", "reparaciones"),
    ("Necesito el contrato de alquiler", "documentacion"),
    ("¿Dónde está el reglamento?", "documentacion"),
    ("Quiero alquilar la unidad 2A", "ofertas"),
    ("¿Hay unidades disponibles?", "ofertas"),
    ("Hola", "general"),
    ("Gracias", "general"),
    ("¿Quién ganó el partido?", "general")
]
# Este df terminaria siendo la base de conocimiento.
df = pd.DataFrame(data, columns=["texto", "categoria"])
df
```

```python
# Entrenar un clasificador ML simple (TF-IDF + LogisticRegression)
# Mantendremos la base de conocimiento (BASE_DE_CONOCIMIENTO)
# y los clasificadores (TF-IDF + LogisticRegression) que ya tienes. Esto es crucial porque la IA generativa puede cometer errores o "alucinar",
# por lo que el clasificador nos ayuda a guiarla hacia la información correcta.

from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LogisticRegression

# Pipeline sencillo
pipeline = Pipeline([
    ("tfidf", TfidfVectorizer()),
    ("clf", LogisticRegression(max_iter=1000))
])

X = df["texto"]
y = df["categoria"]

pipeline.fit(X, y)

# Prueba rápida
test_msgs = [
    "La luz del pasillo no funciona",
    "¿Cómo pago las expensas?",
    "Necesito ver mi contrato",
    "Busco un plomero"
]

for m in test_msgs:
    print(m, "->", pipeline.predict([m])[0])
```

```python
# Generador de respuestas con IA


def generar_respuesta_ia(mensaje_usuario):
    # Usamos el clasificador para guiar a la IA
    categoria = clasificador_combinado(mensaje_usuario)

    # Le damos a la IA un contexto específico basado en la categoría
    contexto_para_ia = ""
    if categoria in BASE_DE_CONOCIMIENTO:
        contexto_para_ia = f"Basado en la siguiente información: {BASE_DE_CONOCIMIENTO[categoria]}. "

    prompt_completo = f"""
    Eres un asistente virtual para la administración de un consorcio. Tu función es responder a las consultas de propietarios, inquilinos o proveedores de manera concisa y útil.

    Instrucciones:
    1. Responde solo con información relevante al consorcio.
    2. Si la pregunta es sobre "reparaciones", solo responde con una frase que confirme la recepción del reclamo y pida más detalles. No inventes soluciones.
    3. Si la pregunta es sobre "pagos" o "documentación", usa la información de la base de conocimiento.
    4. Si la pregunta no se relaciona con el consorcio, responde de manera cortés que solo puedes asistir en temas del edificio.

    Pregunta del usuario: "{mensaje_usuario}"
    """

    response = chat.send_message(prompt_completo)
    return response.text

# Prueba
mensaje_prueba = "¿Cuándo vencen las expensas?"
respuesta_ia = generar_respuesta_ia(mensaje_prueba)
print(respuesta_ia)
```

🛠️ Componentes
Twilio: Una plataforma de comunicación que nos permite enviar y recibir mensajes de WhatsApp y SMS. Actuará como el "teléfono" del bot.

Supabase: Una base de datos relacional de código abierto que nos permitirá almacenar y gestionar datos de manera real, como usuarios, reclamos y la base de conocimiento. También tiene funciones de autenticación y almacenamiento.

API de IA (Google Gemini): El "cerebro" que procesará los mensajes y generará las respuestas.

✅ Lógica y Flujo de Trabajo
El flujo de trabajo es más complejo, ya que implica la integración de múltiples servicios. Aquí te presento los pasos principales:

Configuración en Twilio:

Configura un número de WhatsApp o SMS en tu cuenta de Twilio.

Define una webhook que apunte a tu servidor (por ejemplo, una aplicación web en Flask o FastAPI) que se activará cada vez que un usuario envíe un mensaje a tu número.

Base de datos en Supabase:

Crea un proyecto en Supabase.

Servidor del Chat (por ejemplo, con Flask o FastAPI):

Este será el código principal que ejecutará la lógica.

# Sobre el Prototipo Avanzado


🔹 Metodología

El proyecto se llevará a cabo bajo un enfoque iterativo e incremental, comenzando con un prototipo básico (Gemini API) para validar la capacidad de la IA de generar respuestas útiles, y evolucionando hacia un prototipo avanzado (Twilio + Supabase) que integra comunicación real y almacenamiento persistente.

Los procedimientos implementados incluyen:

Diseño modular: cada componente (clasificador, IA, comunicación, base de datos) se desarrolla de forma independiente para facilitar pruebas y mejoras.

Validación progresiva: primero se prueban respuestas en entorno controlado (Google Colab) y luego se validan en un flujo real con Twilio.

Retrieval-Augmented Generation (RAG): se guía a la IA con información de la base de conocimiento para evitar respuestas incorrectas o inventadas.

✅ Justificación: este enfoque permite reducir riesgos y garantizar resultados tangibles en cada fase, asegurando que los objetivos se cumplen antes de pasar al siguiente nivel.

🔹 Herramientas y Tecnologías

Google Gemini API: motor de IA para generación de respuestas.

Twilio: comunicación por WhatsApp/SMS con usuarios reales.

Supabase: base de datos relacional en la nube, con autenticación y almacenamiento dinámico.

Flask / FastAPI: servidor web para manejar la lógica y conectar servicios.

Python + Scikit-learn: clasificación de mensajes con TF-IDF + Logistic Regression.

dotenv: manejo seguro de claves de API.

Técnicas de Fast Prompting utilizadas

Role prompting: se define claramente el rol de la IA (“asistente virtual de consorcio”).

Instruction prompting: se establecen reglas explícitas (“no inventes soluciones, responde solo sobre el consorcio”).

Context prompting: se aporta información del clasificador y la base de conocimiento como contexto.

Guardrails prompting: se agregan instrucciones de seguridad para evitar respuestas fuera de dominio.

Prompt colaborativo para el desarrollo:

Constraint-Based Prompting: Este método implica establecer reglas y limitaciones claras para la respuesta para el desarrollo de los prototipos.

AI-in-the-loop: (IA en el bucle) o, de forma más específica, "iterative prompting" (prompting iterativo)

Role Prompting: Asignar un rol o persona específica al modelo puede cambiar drásticamente el tono y el estilo de su respuesta. Al indicarle al modelo que actúe como un experto

Few-Shot Prompting: técnica consiste en incluir uno o más ejemplos del formato de entrada y salida deseado dentro del prompt.


**Justificación**: estas técnicas aseguran respuestas relevantes, seguras y consistentes, alineadas con los objetivos del proyecto.

```python
#Implementación

#Ejemplo de código (simplificado, Gemini API)

import os
from dotenv import load_dotenv
import google.generativeai as genai

# Cargar variables de entorno
load_dotenv()
GOOGLE_API_KEY = os.getenv("GOOGLE_API_KEY")

# Configuración de la API
genai.configure(api_key=GOOGLE_API_KEY)
model = genai.GenerativeModel('gemini-pro')
chat = model.start_chat(history=[])

BASE_DE_CONOCIMIENTO = {
    "pagos": "Las expensas vencen el día 10 de cada mes.",
    "documentación": "Los reglamentos pueden solicitarse en administración.",
    "reparaciones": "Debe detallarse el problema para derivar al proveedor correspondiente."
}

def generar_respuesta_ia(mensaje_usuario):
    prompt = f"""
    Eres un asistente virtual de consorcio.
    Solo respondes sobre expensas, documentación o reclamos de reparaciones.
    Pregunta del usuario: "{mensaje_usuario}"
    """
    response = chat.send_message(prompt)
    return response.text

# Prueba
print(generar_respuesta_ia("¿Cuándo vencen las expensas?"))
```

```python
#Ejemplo de código (simplificado, Twilio + Supabase)

from flask import Flask, request
from twilio.twiml.messaging_response import MessagingResponse
from supabase import create_client

app = Flask(__name__)

SUPABASE_URL = os.getenv("SUPABASE_URL")
SUPABASE_KEY = os.getenv("SUPABASE_KEY")
supabase_client = create_client(SUPABASE_URL, SUPABASE_KEY)

@app.route("/sms", methods=['POST'])
def sms_reply():
    mensaje = request.form.get('Body')
    numero = request.form.get('From')

    # Generar respuesta con IA
    respuesta = generar_respuesta_ia(mensaje)

    # Registrar en Supabase
    supabase_client.table("reclamos").insert({
        "telefono": numero,
        "mensaje": mensaje,
        "respuesta": respuesta
    }).execute()

    # Responder al usuario
    resp = MessagingResponse()
    resp.message(respuesta)
    return str(resp)

if __name__ == "__main__":
    app.run(debug=True)
```
